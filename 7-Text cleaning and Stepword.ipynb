## Remove Repeating Characters
### how to simple clean word that have multi duplicate alphabet like loooveee


```python
nltk.download("wordnet")
import re
from nltk.corpus import wordnet

class RepeatReplacer(object):
    
    def __init__(self):
        """
        init of RepeatReplacer
        complie regex (\w*)(\w)\2(\w*)

        """
        self.repeat_regexp=re.compile(r'(\w*)(\w)\2(\w*)')
        self.repl=r'\1\2\3'

    def replace(self,word):
        """
        convert any word into some word like wordnet word
        """
        # if word exsit in wordnet then return it
        if(wordnet.synsets(word)):
            return word
        # complie and use regex on word
        repl_word = self.repeat_regexp.sub(self.repl,word)
        # if word before and after regex now same do it again
        if(repl_word!=word):
            print("****")
            print(repl_word)
            return self.replace(repl_word)
        
        else:
            return repl_word
```

```python
replacer = RepeatReplacer()
replacer.replace("hiiii loooove what's upppppp?")
```

## Removing Stopwords
simple way


```python
from nltk.tokenize import word_tokenize
words = word_tokenize("Here is a good city")
[w for w in words if w not in stopwords.words("english")]
```

### Let's check how much of overheard text is Stopword

```python
from nltk.corpus import webtext
text = webtext.raw("overheard.txt")

text = text.lower()
non_stopwords = [word for word in word_tokenize(text) if word not in stopwords.words("english")]
my_stopwords = [word for word in word_tokenize(text) if word in stopwords.words("english")]
```
